```python
import numpy as np
import pandas as pd
```

```python
d = open("../scraper/raw-data.txt", "r")
data = d.read()
d.close()
len(data)
```

```python
data = data.split('\n')
len(data)
```

```python
market_cap = np.zeros(len(data)-1)
for i in range(len(data)):
    st = data[i].split(" ")[-1]
    st = st.replace("$", "")
    st = st.replace(",", "")
    try:
        market_cap[-(i+1)] = int(st)
    except:
        continue
print(market_cap[:20])
print(market_cap[len(market_cap)-1])
```

```python
#for 1 prediction;
#10 data #11th = prediction
mc11 = np.zeros((market_cap.shape[0]-11+1, 11))
for i in range(market_cap.shape[0]-11+1):
    mc11[i] = market_cap[i:i+11]

print(mc11[:2])
print(mc11[len(mc11)-1])

#20 data #21st = prediction
mc21 = np.zeros((market_cap.shape[0]-21+1, 21))
for i in range(market_cap.shape[0]-21+1):
    mc21[i] = market_cap[i:i+21]

print(mc21[:2])
print(mc21[len(mc21)-1])

#30 data #31st = prediction
mc31 = np.zeros((market_cap.shape[0]-31+1, 31))
for i in range(market_cap.shape[0]-31+1):
    mc31[i] = market_cap[i:i+31]

print(mc31[:2])
```

```python
prefix = 'd'
columns = []
for i in range(10):
    columns.append(prefix+str(i))
columns.append('pred')

df = pd.DataFrame(mc11, columns = columns)
df.to_csv('mc11.csv', index=False)
df
```

```python
prefix = 'd'
columns = []
for i in range(20):
    columns.append(prefix+str(i))
columns.append('pred')

df = pd.DataFrame(mc21, columns = columns)
df.to_csv('mc21.csv', index=False)
df
```
